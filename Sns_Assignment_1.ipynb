```
import numpy as np
import pandas as pd
import math
import cmath
```

```
df=pd.read_csv('signal_assignment1.csv')
df.head()
```

```
df.shape
```

```
x_lis = [i for i in df["x[n]"]] #creating a list of all values of x[n]
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
x = list(np.arange(101))
markerline, stemlines, baseline = plt.stem(x, x_lis, '-.')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
```

```
omega = list(np.linspace(-math.pi, math.pi, 101)) #creating descrete omega between -pi to pi at a constant space 
```

```
x = 0
y = -1
fourier_transform = []
final_omega=[]
z= complex(x,y)  #creating -j
for Ω in omega:
  sum=0
  for n in range(101):
    Y = (cmath.exp(z*Ω*n))*(x_lis[n])       
    sum+=Y
  if abs(sum)>0:
    final_omega.append(Ω)  #omega is added to the list whose dtft is non zero
  fourier_transform.append(abs(sum)) #here dtft for each omega is denoted by sum
print((final_omega))
print((fourier_transform))
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.plot(omega,fourier_transform) #plotting graph between descrete omega and their Dtft
plt.xlabel('Ω')
plt.ylabel('X(e^jΩ)')
plt.show()
```

```
freq=[]
for i in final_omega:
  small_omega=(100*i)    # It is kind ofconverting descrete frequency to continuous frequency
  freq.append(small_omega)
for j in freq:
  if(j>25):
    print('the equipment requires maintenance')
    break
else:
  print('the equipment does not requires maintenance')
print(freq)
  
```

```
plt.plot(final_omega,freq)
plt.xlabel('Ω')
plt.ylabel('ω')
```
